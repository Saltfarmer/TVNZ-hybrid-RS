```python
import pandas as pd
from faker import Faker
import random
```

```python
fake = Faker()
id = []
name = []
username = []
birthdate = []
job = []
sex = []
```

```python
fake.profile()
```

```python
fake.date_of_birth(minimum_age = 18)
```

```python
for i in range(101):
    id.append(i)
    name.append(fake.profile()['name'])
    username.append(fake.profile()['username'])
    birthdate.append(fake.date_of_birth(minimum_age = 18))
    job.append(fake.profile()['job'])
    sex.append(fake.profile()['sex'])
```

```python
len(sex)
```

```python
from datetime import date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
calculate_age(birthdate[6])
```

```python
user_data = pd.DataFrame()
```

```python
user_data['ID'] = id
user_data['name'] = name
user_data['username'] = username
user_data['birthdate'] = birthdate
user_data['job'] = job
user_data['sex'] = sex
user_data
```

```python
user_data.to_csv("user_data.csv", index=False)
```

```python
content_data = pd.read_csv("TVNZ_movies_v3.csv")
```

```python
content_data
```

```python
rating = []
comment = []
user_id = []
content_id = []

for _ in range(15000):
    rating.append(random.randint(1,5))
    comment.append(fake.paragraph(nb_sentences = 10))
    user_id.append(random.randint(0, max(user_data.ID)))
    content_id.append(random.randint(0, max(content_data.ID)))
```

```python
rating_data = pd.DataFrame()
rating_data['rating'] = rating
rating_data['comment'] = comment
rating_data['user_id'] = user_id
rating_data['content_id'] = content_id
rating_data
```

```python
rating_data.to_csv("rating_data.csv", index=False)
```

```python
combine = content_data.merge(rating_data.merge(user_data, left_on=['user_id'], right_on=['ID']), left_on=['ID'], right_on=['content_id'] )
```

```python
combine.to_csv('combine.csv', index=False)
```
